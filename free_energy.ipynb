```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def p_model(x,a,b):
    p0=np.exp(-x**2/2)/(2*np.pi)**0.5
    p1=np.exp(-(x-b)**2/2)/(2*np.pi)**0.5
    return (1-a)*p0+a*p1
```

```python
a0=0.4
b0=4
N=100 #サンプル数
X=[]
for n in range(N):
    u=np.random.random()
    if u<1-a0:
        x=np.random.normal(0,1)
        X.append(x)
    else:
        x=np.random.normal(b0,1)
        X.append(x)
lin=np.linspace(-10,10,100)
plt.plot(lin,p_model(lin,a0,b0))
#plt.hist(X)
plt.show()
```
